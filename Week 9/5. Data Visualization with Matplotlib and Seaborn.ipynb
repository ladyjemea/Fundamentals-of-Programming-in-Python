## Data Visualization with Matplotlib and Seaborn

**Introduction to Matplotlib:**

Matplotlib is a widely-used plotting library for the Python programming language. It provides an object-oriented API for embedding plots into applications and creating static, interactive, and animated visualizations. Matplotlib is highly versatile and is the foundational library for data visualization in the Python ecosystem, often used in conjunction with libraries like NumPy, pandas, and SciPy. 

**Key Features of Matplotlib**
- **Wide Range of Plot Types:**
    - Supports line, bar, scatter, histogram, box, and pie charts.
    - Enables creation of complex visualizations with subplots and custom layouts.
- **Customization:**
    - Extensive options to customize the appearance of plots (colors, labels, scales, legends, etc.).
    - Provides fine control over every aspect of a plot, including fonts, lines, markers, and annotations.
- **Interactive  and Export Options:**
    - Interactive plots for Jupyter Notebooks with zooming and panning.
    - Export plots to multiple formats, including PNG, PDF, and SVG.
- **Integration:**
    - Works seamlessly with libraries like NumPy and pandas for data handling.

#### Common Uses of Matplotlib in Data Science 
- **Exploratory Data Analysis (EDA):** Visualizing distributions, trends, and relationships.
- **Data Presentation:** Creating polished visualizations for reports and presentations.
- **Model Evaluation:** Plotting metrics such as learning curves, ROC curves, or predicted vs actual values.


**Matplotlib Examples:**
- **Basic Plotting**

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Line Plot')
plt.show()
```

**Figure and Axes Objects:**
- Understanding the Figure and Axes:

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 9, 16])
ax.set_xlabel('X-axis Label')
ax.set_ylabel('Y-axis Label')
ax.set_title('Simple Line Plot')
plt.show()
```

**Common Plot Types**
- **Line Plot:**

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

- **Bar Plot:**

```python
plt.bar(['A', 'B', 'C', 'D'], [1, 4, 9, 16])
plt.show()
```

- **Histogram:**

```python
plt.hist([1, 1, 2, 2, 2, 3, 3, 4])
plt.show()
```

- **Scatter Plot:**

```python
plt.scatter([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

**Customizing Plot Appearance**
- **Titles, Labels, and Legends:**

```python
# Customizing plots
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], color='red', marker='o', linestyle='--')
plt.title('Customized Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(['Data'])
plt.annotate('Point', xy=(2, 4), xytext=(3, 5),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

**Matplotlib example**

```python
import matplotlib.pyplot as plt
import numpy as np

# Creating data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Creating a simple line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='sin(x)', color='blue', linestyle='--', linewidth=2)

# Adding title and labels
plt.title('Sine Wave')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Adding a legend
plt.legend()

# Displaying the plot
plt.show()
```

### Seaborn Overview
Seaborn is built on Matplotlib and simplifies creating visually appealing and informative plots. It is particularly effective for statistical graphics and works seamlessly with pandas DataFrames.

**Key Features**
- **Attractive Default Aesthetics:** Built-in themes and color palettes improve the appearance of plots.
- **Simplified Syntax:** High-level functions make it easy to create complex visualizations.
- **Statistical Visualizations:** Specialized functions for regression, distribution, and categorical data.
- **Integration with Pandas:** Directly supports pandas DataFrames for quick plotting.
- **Advanced Plot Types:** Includes heatmaps, pair plots, and facet grids.

**Common Uses of Seaborn**
- **EDA:** Visualize data distributions and relationships.
- **Statistical Analysis:** Regression plots, distribution plots, and confidence intervals.
- **Categorical Data:** Box plots, violin plots, bar plots, and swarm plots.
- **Heatmaps:** Explore correlations and clustered data.

**Seaborn Examples**
- **Basic Plotting**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()
tips = pd.read_csv('data.csv', delimiter=";")

print(tips)

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

**Common Plot Types in Seaborn**
- **Line Plot:**

```python
import seaborn as sns
sns.lineplot(x='total_bill', y='tip', data=tips)
plt.show()
```

- **Scatter Plot:**

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

- **Bar plot:**

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

- **Box Plot:**

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

- **Violin plot:**

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

- **Heatmap:**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation
numeric_tips = tips.select_dtypes(include=['float', 'int'])

# Compute the correlation matrix
corr = numeric_tips.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

**Seaborn vs Matplotlib:**



- Use Matplotlib for advanced control and custom layouts.
- Use Seaborn for quick, attractive plots and statistical graphics.
