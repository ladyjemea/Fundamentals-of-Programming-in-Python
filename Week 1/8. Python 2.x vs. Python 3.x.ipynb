# Python 2 vs. Python 3

Python 2 and Python 3 are two major versions of Python, and while Python 2 was popular for many years, Python 3 is now the standard. Python 2 has officially reached end of life (EOL), meaning it is no longer supported or updated. Here's a comparison between the two versions:

1. **Print Statement:** <br>
- In **Python 2**, print is treated as a statement, not a function. You can write:


- In **Python 3**, print is a function, so you need parentheses:

```python
print("Hello, World!")
```

2. **Integer Division:**
- In **Python 2**, dividing two integers results in integer division. For example:


- In **Python 3**, dividing two integers results in a float:

```python
print(5 / 2)  # Outputs: 2.5
```

- To get integer division in Python 3, you need to use //:

```python
print(5 // 2)  # Outputs: 2
```

3. **Unicode Support:**
- **Python 2** uses ASCII as the default encoding for strings.
- **Python 3** uses Unicode by default, making it much better suited for applications involving multiple languages and special characters. <br>

Example in **Python 3**:

```python
text = "Hello, 世界"  # Supports non-ASCII characters by default
```

4. **Input Function:**
In **Python 2**, the **input()** function evaluates the input as Python code, while raw_input() is used to get a string.

In **Python 3**, input() always returns a string:

```python
name = input("Enter your name: ")  # Python 3
```

5. **Libraries:**
Many new third-party libraries and features are only compatible with Python 3. While Python 2 libraries are still available, future developments are focused on Python 3.

6. **End of Life (EOL):**
**Python 2** reached its end of life on **January 1, 2020**. There will be no more official updates or security fixes for Python 2, so all **new development** should be done in **Python 3**.

7. **Range Function:**
- In **Python 2**, range() returns a list, which can consume a lot of memory for large ranges.
- In Python 3, range() returns an immutable sequence object (similar to a generator), making it more memory efficient.
