# Installing Python

There are several ways to run Python code. For this course, we will focus on three options, though we will primarily use two of them:

- Installing Python with the **Anaconda** distribution.
- Downloading **Python** locally on your computer.
- No-install options for running Python code.

Let’s dive into these options in more detail.

# Option 1: Installing Python with the Free Anaconda Distribution

The Anaconda distribution is a free and open-source platform that includes both Python and the Jupyter Notebook, which we will use throughout this course for writing and running code. Anaconda makes Python installation easy and includes many of the tools you’ll need. It works on any major operating system: Windows, MacOS, and Linux.

### Steps to Install Anaconda
1) **Go to the Anaconda Downloads Page:** <br>
Visit www.anaconda.com/downloads.

2) **Choose Your Operating System:** <br>
Select the appropriate installer for your operating system (Windows, MacOS, or Linux). <br>
- Download the 64-bit graphical installer for your system.

3) **Follow the Installation Guide:** <br>
For step-by-step instructions specific to your operating system, check out Anaconda's official installation guide: docs.anaconda.com/anaconda/install.

4) **Run the Installer:** <br>
- Once downloaded, open the installer. It's usually best to accept the default installation settings.
- For Windows Users: If this is your first Python installation, check the box that says **"Add Anaconda to PATH"**. This will make sure that your system recognizes Anaconda as the default Python environment.

5) **Finish Installation:** <br>
Complete the installation by following the on-screen instructions. It’s fine to accept all default settings.


### Running Anaconda and Jupyter Notebook

1) **Open Anaconda Navigator:** <br>
Once installed, search your computer for Anaconda Navigator and open it.

2) **Launch Jupyter Notebook:** <br>
In Anaconda Navigator, you’ll see various tools and environments.
- Click **Launch** under **Jupyter Notebook**. This will open a Jupyter session in your default web browser.

3) **Navigating Jupyter:**
Jupyter launches in your main user directory, where you can see all your files and folders. You can navigate through your files or create new ones directly from the Jupyter interface.
- To create a new notebook, click **New -> Notebook -> Python 3**. This opens a new Python notebook where you can write and execute Python code.

### What is Jupyter Notebook?

**Jupyter Notebook** is an interactive environment where you can write and execute Python code in cells. It’s widely used for documentation, data analysis, machine learning, and more. Each "cell" can contain text, code, or visualizations, making it ideal for beginners who want to experiment and see immediate feedback.

- To run a cell, press **Shift + Enter.**
- You can save your work, add notes, and see results instantly, which is great for learning!

### Closing and Managing Notebooks
- When you're done, simply close the notebook tab in your browser and make sure to shut down the notebook from the Jupyter interface.
- You can create, delete, move, or download notebooks. Jupyter saves your progress automatically, and files are saved with the .ipynb extension (not the usual .py Python script format).

# Option 2: Downloading Python Locally on Your Computer

If you prefer not to use Anaconda, you can install Python directly on your computer.

### Steps to Install Python
1) **Go to the Official Python Website:** <br>
Visit python.org/downloads.

2) **Download the Latest Version:** <br>
Download the latest version of Python for your operating system (Windows, MacOS, or Linux).

3) **Run the Installer:** <br>
Open the downloaded file and follow the installation instructions.
- **For Windows Users:** Make sure to check the box that says “**Add Python to PATH**” during installation. This will allow you to run Python from the command prompt.

4) **Complete the Installation:** <br>
Accept the default settings, and click Install. Once installed, Python will be ready to use.

### Testing Your Python Installation
1) **Open the Command Prompt or Terminal:** <br>
- For Windows, search for **Command Prompt**.
- For MacOS/Linux, open the **Terminal**.

2) **Check Python Installation:** <br>
- Type **python --version** (or **python3 --version** on some systems) and press Enter. You should see the version of Python installed. If Python was installed correctly, you’ll see the version number displayed.
- If everything is set up properly, you can start writing simple Python commands in the command prompt or terminal.

### Installing a Code Editor (Visual Studio Code)
A code editor makes it easier to write and manage Python code. For this course, we recommend Visual Studio Code (VS Code), a popular, lightweight editor with Python support.

#### Steps to Install Visual Studio Code
1) **Go to the VS Code Website:** <br>
Visit code.visualstudio.com.

2) **Download and Install VS Code:** <br>
Download the latest version of VS Code for your operating system and follow the installation instructions.
- For Windows Users: Make sure to check **"Add to PATH"** during installation.

3) **Set Up Python in VS Code:**
- Once VS Code is installed, open the Extensions tab (left sidebar) and search for Python.
- Install the **Python extension by Microsoft** to enable Python support in VS Code.

4) **Create a Workspace:**
- On your computer, create a folder called **Python Fundamentals**. This is where you’ll save all your Python projects for this course.
- Open this folder in VS Code, and you’re ready to start writing Python code!

# Option 3: No-Install Options to Run Python Code

If you're on the go and don’t have access to your installed Python setup, there are no-install options that let you run Python code directly from your browser. Although we’ll primarily use Jupyter Notebooks or locally installed Python for this course, here are a few web-based Python environments:

1) **Repl.it:** <br>
A popular online code editor that supports Python. Visit **replit.com** and start coding without any setup.

2) **Jupyter.org/try:** <br>
Run Jupyter notebooks directly from your browser. Visit **try.jupyter.org**.

3) **Google Colab:** <br>
Google's cloud-based platform for running Python code, especially useful for data science tasks. You’ll need a Google account to access **Google Colab**.

4) **W3Schools Python Editor:** <br>
A simple Python interpreter available on **W3Schools**, great for testing small code snippets. <br>

These platforms are useful for quick experimentation or collaboration when you don’t have access to your computer.
