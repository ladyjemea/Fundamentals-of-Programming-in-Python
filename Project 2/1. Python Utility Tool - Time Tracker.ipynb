**Project Description:**
<br>
Develop a Python program that helps users track and manage their daily tasks and activities. The program will allow users to add tasks, set timers, view logs of completed tasks, and calculate total time spent on different activities. This project integrates concepts like functions, loops, conditionals, file handling, and error handling.
<br><br>
**Requirements:**
<br>
- **Task Management:**
Allow users to add new tasks with descriptions and optional priorities (e.g., High, Medium, Low).
View a list of ongoing tasks.

- **Timer Functionality:**
Let users start and stop a timer for a specific task.
Record the start and end times for each task session.

- **Task Log:**
Save completed tasks with timestamps and total time spent to a log file (e.g., task_log.txt).
Display a summary of completed tasks with time spent.

- **Total Time Calculation:**
Calculate the total time spent on all tasks or specific tasks.

- **Error Handling:**
Handle invalid inputs, such as attempting to stop a timer that hasn’t been started.
Ensure the program gracefully manages unexpected issues, such as missing files.

- **Menu-Driven Interface:**
    Present a menu for users to navigate options like:
    - Add Task
    - View Tasks
    - Start/Stop Timer
    - View Completed Tasks
    - Exit
<br>

**Evaluation Criteria:**
- **Functionality:** Does the program accurately track time and manage tasks?

- **Code Quality:** Is the code modular, well-structured, and appropriately commented?

- **Error Handling:** Does the program handle invalid inputs and unexpected situations gracefully?

- **User Experience:** Is the menu intuitive and easy to use?


**Deliverables:**
- **Code:**
    - A Python script implementing the time tracker.
- **Documentation:**
    - Step-by-Step Explanation: A detailed walkthrough of the code.
    - Usage Instructions: How to set up and use the program.
    - Feature Explanation: Description of the program’s features and functionality.


**Sample Features Integrating Class Concepts:**

- **Loops and Conditionals:**
Use loops for the menu and iterating over tasks.
Apply conditionals to check the state of tasks (e.g., whether a timer is running).

- **Functions:**
Modularize operations like adding tasks, starting/stopping timers, and calculating time.

- **File Handling:**
Save completed task logs to a file and retrieve them when needed.

- **Error Handling:**
Handle issues like stopping a timer for a non-existent task.

- **Modules:**
Use Python’s built-in modules (time, os) for timing and file operations.
