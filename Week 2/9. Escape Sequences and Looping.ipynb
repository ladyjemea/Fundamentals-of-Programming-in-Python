# Escape Sequences in Strings
Escape sequences are special characters used in strings to represent non-printable or illegal characters. They begin with a backslash (\) followed by the character(s) you want to insert. They are useful for including special formatting or non-printable characters within strings.

### Common Escape Sequences:


**Examples:**

```python
print("Hello World")
print("Hello\nWorld")   # New line
print("Hello\tWorld")   # Tab space
print("This is a backslash: \\")  # Backslash
print("\x48")           # ASCII code for 'H'
```

### Looping Through Strings

Looping through strings allows you to iterate over each character in the string. This can be useful for performing operations on individual characters or for processing the string as a sequence.

1. **Looping with for Loop** <br>
You can loop over each character in a string directly using a for loop:

**Example:**

```python
text = "Python"
for char in text:
    print(char)
```

2. **Looping with the range() Function** <br>
Using the range() function, you can iterate over the indices of the string and use indexing to access each character.

**Example:**

```python
text = "Python"
for i in range(len(text)):
    print(text[i])
```

3. **Looping with the enumerate() Function** <br>
The enumerate() function allows you to iterate over both the indices and the characters of the string simultaneously. This is especially useful when you need to keep track of the index while looping.

**Example:**

```python
text = "Python"
for index, char in enumerate(text):
    print(f"Index: {index}, Character: {char}")
```
