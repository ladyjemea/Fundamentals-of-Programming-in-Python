# Python Operators

Operators are symbols or keywords used to perform operations on variables and values. They allow you to manipulate data, perform calculations, compare values, and control the flow of a program. Python supports several types of operators:

1) Arithmetic Operators
2) Assignment Operators
3) Comparison Operators
4) Logical Operators
5) Identity Operators
6) Membership Operators
7) Bitwise Operators

1. **Arithmetic Operators** <br>
Arithmetic operators are used for performing mathematical operations on numeric values.



**Example:**

```python
x = 10
y = 2

# Addition
result = x + y  # result = 12

# Subtraction
result = x - y  # result = 8

# Multiplication
result = x * y  # result = 20

# Division
result = x / y  # result = 5.0

# Modulus (remainder)
result = x % y  # result = 0

# Exponentiation (power)
result = x ** y  # result = 100

# Floor division
result = x // y  # result = 5
```

2. **Assignment Operators** <br>
Assignment operators are used to assign values to variables and can also perform an operation while updating the variable.



**Example:**

```python
x = 10

# Add and assign
x += 3  # x = 13

# Subtract and assign
x -= 3  # x = 10

# Multiply and assign
x *= 3  # x = 30
```

3. **Comparison Operators** <br>
Comparison operators are used to compare two values. The result is always a Boolean (True or False).



**Example:**

```python
x = 10
y = 2

print(x == y)  # Output: False
print(x != y)  # Output: True
print(x > y)   # Output: True
```

4. **Logical Operators** <br>

Logical operators are used to combine conditional statements and return a Boolean result (True or False).



**Example**

```python
x = 10

print(x > 5 and x < 20)  # Output: True
print(x > 5 or x < 9)    # Output: True
print(not(x > 5))        # Output: False
```

5. **Identity Operators** <br>

Identity operators are used to compare objects to check if they are the same object in memory.



**Example:**

```python
x = [1, 2, 3]
y = [1, 2, 3]
z = x

print(x is y)  # Output: False (different objects)
print(x is z)  # Output: True (same object)
print(x == y)  # Output: True (same content)
```

6. **Membership Operators** <br>

Membership operators are used to check if a value exists in a sequence (like lists, tuples, or strings).



**Example:**

```python
x = [1, 2, 3]

print(3 in x)      # Output: True
print(4 not in x)  # Output: True
```

7. **Bitwise Operators** <br>

Bitwise operators perform operations at the bit level. They compare individual bits in binary representations of integers.



**Example:**

```python
x = 5    # 0101 in binary
y = 3    # 0011 in binary

print(x & y)  # Output: 1  (AND operation)
print(x | y)  # Output: 7  (OR operation)
print(x ^ y)  # Output: 6  (XOR operation)
```

### Operator Precedence

Operator precedence determines the order in which operations are performed. Operators with higher precedence are evaluated first. Parentheses can be used to override precedence.
